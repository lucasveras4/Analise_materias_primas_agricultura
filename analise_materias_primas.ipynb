```python
# Importando as bibliotecas requeridas
import numpy as np
import pandas as pd
```

**Carregando a base de dados em um dataframe utilizando pandas**

```python
# usando o método read_csv()
df = pd.read_csv('agricultural_raw_material.csv')
```

```python
# Visualização do dataframe
df
```

**Explorando o banco de dados**

```python
# Checando valores nulos em cada coluna
df.isnull().sum()
```

**Tratando dados ausentes, incorretos e inválidos**

```python
# Substituindo valores "%", "," e "-" para valores em branco
df = df.replace('%', '', regex=True)
df = df.replace(',', '', regex=True)
df = df.replace('-', '', regex=True)
df = df.replace('', np.nan)
df = df.replace('MAY90', np.nan)

# Dropando linhas com valores NaN (not a number)
df = df.dropna()

# Checando novamente os valores nulos
df.isnull().sum()
```

```python
df1 = df
```

```python
# Conertendo o datatype de int64 para float64
lista = ['Coarse wool Price','Coarse wool price % Change','Copra Price','Copra price % Change','Cotton Price','Cotton price % Change','Fine wool Price','Fine wool price % Change','Hard log Price','Hard log price % Change','Hard sawnwood Price','Hard sawnwood price % Change','Hide Price','Hide price % change','Plywood Price','Plywood price % Change','Rubber Price','Rubber price % Change','Softlog Price','Softlog price % Change','Soft sawnwood Price','Soft sawnwood price % Change','Wood pulp Price','Wood pulp price % Change']
df[lista] = df[lista].astype('float')
df.dtypes
```

```python
df.head()
```

```python
# Mudando a formatação da data na coluna 'Month'
df.Month = pd.to_datetime(df.Month.str.upper(), format='%b%y', yearfirst=False)

# Indexando a coluna Month
df = df.set_index('Month')
```

```python
df.head()
```

**Analise exploratoria e visualização**

```python
# Vamos começar importando as bibliotecas matplotlib.pyplot e seaborn
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9,5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'

```

**Mapa de calor**

```python
# Lista de materias primas
materia_prima = ['Coarse wool Price','Copra Price','Cotton Price','Fine wool Price','Hard log Price','Hard sawnwood Price',
                 'Hide Price','Plywood Price','Rubber Price','Softlog Price','Soft sawnwood Price','Wood pulp Price']

# Obtendo a matriz correlacional
corrmat = df[materia_prima].corr()

# Definindo o tamanho do plot
fig = plt.figure(figsize = (12, 9))
mask = np.triu(np.ones_like(corrmat, dtype=bool))
sns.heatmap(corrmat, vmax= .8, mask=mask, square=True, annot=True)
plt.show()
```

O mapa de calor mostra a correlação entre matérias-primas, quanto mais próximo de 1 o valor for, maior a chance dos valores dessas matérias-primas estarem relacionados, mas não necessariamente.


```python
# Agora vamos fazer o mapa com a variação em % dos valores

variacoes = ['Coarse wool price % Change','Copra price % Change','Cotton price % Change','Fine wool price % Change','Hard log price % Change','Hard sawnwood price % Change',
            'Hide price % change','Plywood price % Change','Rubber price % Change','Softlog price % Change','Soft sawnwood price % Change','Wood pulp price % Change']

# Matriz correlacional
plt.figure(figsize=(30,15))
corrMatrix = df[variacoes].corr()
sns.heatmap(corrMatrix, annot=True)
plt.show()
```

A partir deste gráfico, podemos ver que praticamente não há relação entre a variação percenrual dos preços das matérias-primas

**.** O valor negativo em correlações diz que duas variáveis estão negativamente relacionadas (um aumenta, outro decresce) <br>
**.** Zero implica em nenhuma relação <br>
**.** Caso contrário, quanto mais próximo de 1, maior a a chance relação

```python
# Agora vamos fazer uma analíse individual para as matérias-primas, usando o preço e a variação em % ao longo do tempo com um grafico evolutivo

axes = df[['Coarse wool Price','Coarse wool price % Change' ]].plot(figsize=(11,9), subplots=True, linewidth=1)
```

Da mesma forma, podemos encontrar a visualização de todos os outros recursos fornecidos no conjunto de dados

**Descobrindo a variação do preço de cada matéria-prima**

```python
for i in range(len(variacoes)):
  plt.figure(figsize=(12,12))
  df[variacoes[i]].hist(figsize=(11,9), linewidth=1)
  plt.xlabel('% change')
  plt.ylabel('count')
  plt.legend(variacoes[i:], loc='upper center', bbox_to_anchor=(1.2,1))
```

Podemos observar que a maioria das matérias-primas tem % de mudança frequente ideal inferior a 5%


**Encontrando a matéria-prima que tem o menor preço ao longo dos anos**

```python
plt.figure(figsize=(12,12))

for i in range(len(materia_prima)):
  plt.subplot(4,3,i+1)
  plt.subplots_adjust( hspace=1, wspace=0.5)
  plt.title(materia_prima[i])
  plt.plot(df[materia_prima[i]])
  plt.xticks(rotation=90)
plt.suptitle('Comparação de preços das matérias-primas ao longo dos anos')
```

Podemos ver que o algodão e a borracha são as matérias-primas com os preços mais baixos, vamos compara-los para entender melhor qual tem o preço mais baixo

```python
# Comparando o preço do algodão e borracha
plt.figure(figsize=(10,10))
plt.plot(df[['Cotton Price', 'Rubber Price']])
plt.title('Comparação de preços das matérias-primas')
plt.xlabel('Years')
plt.ylabel('Prices')
plt.legend(['Cotton Price', 'Rubber Price'], loc='upper center', bbox_to_anchor=(1.2,1))
```

Podemos observar que o **Algodão** é a matéria-prima com o menor preço nos últimos anos

**Quais matérias-primas possuem o menor e maior valor de variação % de preço**

```python
import random
```

```python
plt.figure(figsize=(12,12))
for i in range(len(variacoes)):
  r = random.random()
  g = random.random()
  b = random.random()
  color = (r, g, b)
  plt.subplot(4,3,i+1)
  plt.subplots_adjust( hspace=1, wspace=0.5)
  plt.plot(df[variacoes[i]], c=color)
  plt.xticks(rotation=90)
  plt.title(variacoes[i])
  plt.xlabel('Years')
  plt.ylabel('% Change')
```

Podemos ver o que o maior percentual de variação foi na ***madeira serrada***, chegando a mais de 60% de variação, e o menor foi na ***madeira compensada***, com uma variação menor do que 20%
